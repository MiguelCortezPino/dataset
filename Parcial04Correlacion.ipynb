```python
# -------------------------------------------------------------------------
# PASO 2 correlacion - BLOQUE 1: Carga, Identificación de Variables y Limpieza
# -------------------------------------------------------------------------
import pandas as pd
import numpy as np

# Ruta de tu dataset de correlación en la nube en GitHub
url_corr = "https://raw.githubusercontent.com/MiguelCortezPino/dataset/refs/heads/main/Parcial04/clave_H_correlacion.csv"

# Cargamos el archivo CSV directamente usando la URL
df_corr = pd.read_csv(url_corr)

# 1. Mostrar las primeras 5 filas de manera compacta para comprobar que cargó bien
print("--- PRIMERAS FILAS DEL DATASET (Métricas de Clientes) ---")
print(df_corr.head(5))

# 2. Identificar y listar las variables numéricas del dataset (lo pide el examen)
variables_numericas = df_corr.select_dtypes(include=[np.number])
print("\n--- VARIABLES NUMÉRICAS IDENTIFICADAS ---")
print(variables_numericas.columns.tolist())

# 3. Revisar si existen datos nulos en el dataset
print("\n--- REVISIÓN DE DATOS NULOS POR COLUMNA ---")
print(df_corr.isnull().sum())

# Tratamiento de nulos con la Mediana (Limpieza para asegurar consistencia general)
df_corr['satisfaccion'] = df_corr['satisfaccion'].fillna(df_corr['satisfaccion'].median())
df_corr['tiempo_respuesta'] = df_corr['tiempo_respuesta'].fillna(df_corr['tiempo_respuesta'].median())
```

```python
# -------------------------------------------------------------------------
# PASO 2 - BLOQUE 2: Matriz de Correlación y Gráfico de Calor
# -------------------------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt

# 4. Calcular la matriz de correlación lineal de Pearson (solo variables numéricas)
matriz_correlacion = df_corr.select_dtypes(include=[np.number]).corr()

# Mostrar la correlación enfocada en el 'consumo_total' de forma corta y ordenada
print("\n--- MATRIZ DE CORRELACIÓN ENFOCADA EN EL CONSUMO TOTAL ---")
print(matriz_correlacion['consumo_total'].sort_values(ascending=False).round(2))

# 5. Crear el Mapa de Calor (Heatmap) de forma nítida y profesional
sns.set_theme(style="white")
plt.figure(figsize=(10, 8))

# Graficamos la matriz completa con números pequeños bien distribuidos
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)

# Añadimos el título formal
plt.title('Mapa de Calor: Matriz de Correlación Lineal (Clave H)', fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.show()
```
